Coding the above in more dynamic way

````python

weights=[[0.2,0.8,-0.5,1.0],[0.5,-0.91,0.26,-0.5],[-0.26,-0.27,0.17,0.87]]
biases=[2,3,0.5]
inputs=[1,2,3,2.5]
def dynamic_outputs(weights,biases,inputs):
  layer_outputs=[]

  for i_weight,bias in zip(weights,biases):
    n_output=0

    for w,input in zip(i_weight,inputs):
      n_output+=w*input
    output=n_output+bias
    layer_outputs.append(output)
  return layer_outputs






````

````python
import numpy as np
import math
import nnfs
from nnfs.datasets import spiral_data
nnfs.init()
class Hidden_Layers:
  def __init__(self,n_features,n_neurons):
    self.weights=np.random.randn(n_features,n_neurons)*0.1
    self.biases=np.zeros((1,n_neurons))
  def forward(self,inputs):
    self.output=np.dot(inputs,self.weights)+self.biases
    return self.output
  def activation_relu(self,input):
    self.output=np.maximum(0, input)
    return self.output
  def activation_softmax(self,input):
    exp_vals=np.exp(input-np.max(input,axis=1,keepdims=True))
    self.output=exp_vals/np.sum(exp_vals,axis=1,keepdims=True)
    return self.output


````

````python
dense=Hidden_Layers(2,3)
x=dense.forward(X)
relu=dense.activation_relu(x)
softmax=dense.activation_softmax(relu)
print(softmax)
````

````python
import numpy as np
np.random.seed(0)
import math
import nnfs
from nnfs.datasets import spiral_data
nnfs.init()
class Hidden_Layers:
  def __init__(self,n_features,n_neurons):
    self.weights=np.random.randn(n_features,n_neurons)*0.01
    self.biases=np.zeros(n_neurons)
  def forward(self,inputs):
    self.outputs=np.dot(inputs,self.weights)+self.biases
class Activation_Function():
  def relu_activation(self,input):
    self.relu_output=np.maximum(0,input)
  def activation_softmax(self,input):
    exp_vals=np.exp(input-np.max(input,axis=1,keepdims=True))
    self.softmax_output=exp_vals/np.sum(exp_vals,axis=1,keepdims=True)
class Loss:
  def calculate(self,output,y):
    sample_losses=self.forward(output,y)
    data_loss=np.mean(sample_losses)
    return data_loss
class Cross_Entropy(Loss):
  def forward(self,y_pred,y_true):
    clipped_output=np.clip(y_pred,1e-7,1-1e-7)# to make 0s to 0.000001 or nearnest no. cos log(0) gives error
    samples=len(y_pred)
    if len(y_true.shape)==1:
      correct_confidences=clipped_output[range(samples),y_true] # to handle data if thethe true values are in scalar form

    elif len(y_true.shape)==2:
      correct_confidences=np.sum(clipped_output*y_true,axis=1) # to handle daata if the y true values are in OHE form
    output=-np.log(correct_confidences)
    return output

````

````python
X,y=spiral_data(100,3)
````

````python
layer_1=Hidden_Layers(2,3)
layer_1.forward(X)
print(layer_1.outputs)
````

````python
act=Activation_Function()
act.relu_activation(layer_1.outputs)
print(act.relu_output)
````

````python
act.activation_softmax(act.relu_output)
print(act.softmax_output)
````

````python
loss=Cross_Entropy()

y_pred = act.softmax_output
y_true = y
print(loss.calculate(y_pred, y_true))
````
