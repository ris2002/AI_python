Model Architecture



```python
import numpy as np
# hidden layer with activation function
class Hidden_Layer:
  def __init__(self,vocab_size,embed_size,hidden_size):
    #hidden_size is nothing but no. neurons
    self.bias=np.zeros((1,hidden_size))
    #Embeddingtable
    self.Wex=np.random.randn(vocab_size,embed_size)*0.01
    #Hidden-Hidden weights
    self.Whh=np.random.randn(hidden_size,hidden_size)*0.01
    # Embed-Hidden weights
    self.Wxh=np.random.randn(embed_size,hidden_size)*0.01

  def forward(self,X_idx,h0=None):
    T=len(X_idx)
    H=self.Whh.shape[0]
    if h0 is None:
      h_prev=np.zeros((1,H))
    else:
      h_prev=h0.reshape(1,H)
    hs=np.zeros((T, H))
    self.cache=[]
    for t in range(T):
      idx=X_idx[t]
      e_t=self.Wex[idx:idx+1, :]
      a_t=e_t@self.Wxh+h_prev@self.Whh+self.bias
      h_t=a_t
      hs[t]=h_t
      self.cache.append((idx, e_t, h_prev, a_t, h_t))
      h_prev=h_t
    return hs
  def backward(self, dinputs):
    self.dWxh  = np.zeros_like(self.Wxh)
    self.dWhh  = np.zeros_like(self.Whh)
    self.dbias = np.zeros_like(self.bias)
    dh_next    = np.zeros((1, self.Whh.shape[0]))

    for i in reversed(range(len(self.cache))):
        idx, e_t, h_prev, a_t, h_t = self.cache[i]

        dh = dinputs[i] + dh_next

        self.dWxh  += e_t.T    @ dh
        self.dWhh  += h_prev.T @ dh
        self.dbias += np.sum(dh, axis=0, keepdims=True)

        dh_next = dh @ self.Whh.T

    return self.dWxh, self.dWhh, self.dbias


class Output_Layer():
  def __init__(self,hidden_size,vocab_size):
    self.weights=np.random.randn(hidden_size,vocab_size)*0.01
    self.bias=np.zeros((1,vocab_size))
  def forward(self,hidden_state):
    self.inputs=hidden_state
    self.output=np.dot(self.inputs,self.weights)+self.bias
    return self.output
  def backward(self,dinputs):

    self.dweights = self.inputs.T @ dinputs
    self.dbias = np.sum(dinputs, axis=0, keepdims=True)
    do_dh=self.weights.T
    op = dinputs @ do_dh
    return op
class Tanh_Activation_Function():
  def forward(self,inputs):
    self.outputs=np.tanh(inputs)
    return self.outputs
  def backward(self,dinputs):
    tanh_z=self.outputs
    dtanh_dzt=(1-(tanh_z**2))
    op=dtanh_dzt*dinputs
    return op

class Softmax_Layer:
  def forward(self,inputs):
    shifted = inputs - np.max(inputs, axis=1, keepdims=True)
    exp_vals=np.exp(shifted)
    self.output=exp_vals/np.sum(exp_vals,keepdims=True,axis=1)
    return self.output
  def backward(self,dvalues):
    self.dinputs = np.empty_like(dvalues)
    for index, (single_output, single_dvalue) in enumerate(
            zip(self.output, dvalues)
        ):
      single_output = single_output.reshape(-1, 1)
      jacobian_matrix = (
                np.diagflat(single_output)
                - np.dot(single_output, single_output.T)
            )
      self.dinputs[index] = np.dot(jacobian_matrix, single_dvalue)
      return self.dinputs
class Categorical_Cross_Entropy:
  def forward(self,y_pred,y_true):
    clipped_val=np.clip(y_pred,1e-7, 1 - 1e-7)
    samples = len(y_pred)
    clipped_vals=np.clip(y_pred,1e-7, 1 - 1e-7)
    if len(y_true.shape)==1:
      correct_confidences=clipped_vals[range(samples),y_true]
    elif len(y_true.shape)==2:
      correct_confidences=np.sum(clipped_vals * y_true, axis=1)
    data_loss = -np.log(correct_confidences)
    output=np.mean(data_loss)
    return output
  def backward(self,y_pred,y_true):
    samples = len(y_pred)
    labels=len(y_pred[0])
    #np.eye(n) creates an identity matrix with diamenntios n*n
    if len(y_true.shape) == 1:
      y_true=np.eye(labels)[y_true]
      self.d_values=-y_true/y_pred
      self.d_values=self.d_values/samples
      return self.d_values

class SGD:
    def __init__(self, learning_rate=0.01):
        self.lr = learning_rate

    def update(self, layer):
        layer.weights -= self.lr * layer.dweights
        layer.biases -= self.lr * layer.dbiases

```

```python
for epoch in range(100):
    # forward
    a = hidden.forward(X)
    b = tanh.forward(a)
    c = output_layer.forward(b)
    d = softmax.forward(c)
    e = loss.forward(d, np.array(Y))

    # backward
    d1 = loss.backward(d, np.array(Y))
    d2 = softmax.backward(d1)
    d3 = output_layer.backward(d2)
    d4 = tanh.backward(d3)
    dWxh, dWhh, dbias = hidden.backward(d4)

    # update weights
    output_layer.weights -= 0.01 * output_layer.dweights
    output_layer.bias    -= 0.01 * output_layer.dbias
    hidden.Wxh           -= 0.01 * dWxh
    hidden.Whh           -= 0.01 * dWhh
    hidden.bias          -= 0.01 * dbias

    if epoch % 10 == 0:
        print(f"Epoch {epoch} Loss: {e}")
```

```python
import torch
import torch.nn as nn
```

```python
class Vanilla_RNN(nn.Module):
  def __init__(self, vocab_size, embed_size, hidden_size):
    super(Vanilla_RNN,self).__init__()
    self.rnn=nn.RNN(embed_size,hidden_size,batch_first=True)
    self.embed=nn.Embedding(vocab_size,embed_size)
    self.output=nn.Linear(hidden_size, vocab_size)
  def forward(self,x,h0=None):
    e=self.embed(x)
    out,hn=self.rnn(e.unsqueeze(0), h0)
    logits=self.output(out.squeeze(0))
    return logits,hn


```

```python
model     = Vanilla_RNN(vocab_size=27, embed_size=10, hidden_size=100)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(100):
    optimizer.zero_grad()
    x      = torch.tensor(X)
    y      = torch.tensor(Y)
    logits, _ = model(x)
    loss   = criterion(logits, y)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch} Loss: {loss.item()}")
```

```python
def predict(model, start_char, char_to_ix, ix_to_char, max_length=20):
    model.eval()

    # start with given character
    x = torch.tensor([char_to_ix[start_char]])
    h = None
    result = [start_char]

    with torch.no_grad():
        for _ in range(max_length):
            logits, h = model(x, h)

            # get predicted next character
            probs = torch.softmax(logits[-1], dim=0)
            next_idx = torch.multinomial(probs, 1).item()
            next_char = ix_to_char[next_idx]

            # stop if newline
            if next_char == '\n':
                break

            result.append(next_char)
            x = torch.tensor([next_idx])

    return ''.join(result)

# generate names
for _ in range(5):
    print(predict(model, 'a', char_to_ix, ix_to_char))
```
